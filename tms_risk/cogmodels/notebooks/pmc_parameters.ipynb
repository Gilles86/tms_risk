```python
bids_folder = '/data/ds-tmsrisk/'
import os.path as op
import arviz as az

from tms_risk.cogmodels.fit_model import build_model, get_data
import pymc as pm
from tqdm.notebook import tqdm
import pandas as pd

from tms_risk.cogmodels.utils import plot_ppc

from IPython.display import display
```

```python
dfs = []
models = []
idatas = []
models = []

model_labels  = ['10_null', '10a', '10b', '10c', '11_null', '11a', '11b', '11c']


for model_label in tqdm(model_labels):
    dfs.append(get_data(model_label=model_label))

    model = build_model(model_label, dfs[-1])
    model.build_estimation_model()
    idata = az.from_netcdf(op.join(bids_folder, 'derivatives', 'cogmodels', f'model-{model_label}_trace.netcdf')).sel(draw=slice(None, None, 10))

    idatas.append(idata)
    models.append(model)

    
```

```python
conditions = pd.DataFrame([{'stimulation_condition':'vertex'},
                            {'stimulation_condition':'ips'}])


for model_label, model, df, idata in zip(model_labels, models, dfs, idatas):

    pars = model.get_conditionwise_parameters(idata, conditions)
    diff = pars['ips'] - pars['vertex']
    print(model_label)
    display(diff.groupby('parameter').apply(lambda x: (x> 0.0).mean()))
```
