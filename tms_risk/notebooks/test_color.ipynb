```python
import seaborn as sns

# Get the current palette colors
current_palette = sns.color_palette()

# Known Seaborn palettes to compare
palettes = {
    "deep": sns.color_palette("deep"),
    "muted": sns.color_palette("muted"),
    "colorblind": sns.color_palette("colorblind"),
    "pastel": sns.color_palette("pastel"),
    "dark": sns.color_palette("dark"),
    "bright": sns.color_palette("bright")
}

# Function to find the matching palette name
def find_palette_name(current_palette):
    for name, palette in palettes.items():
        if current_palette == palette:
            return name
    return "unknown"

# Print the detected palette name
palette_name = find_palette_name(current_palette)
print("Current palette name:", palette_name)
```
